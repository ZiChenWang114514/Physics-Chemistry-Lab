```python
# 0.01900 -0.0200 -0.0180 -0.0110
class Sample():
    def __init__(self, name, m, m3, m4, m31, m41):
        self.name = name
        self.m = round(m,4)
        self.m3 = round(m3,4)
        self.m4 = round(m4,4)
        self.m31 = round(m31,4)
        self.m41 = round(m41,4)
        

    def __str__(self):
        return self.name + ' ' + str(self.m) + ' ' + str(self.m3) + ' ' + str(self.m4) + ' ' + str(self.m31) + ' ' + str(self.m41)
    __repr__ = __str__

# 取data2第一行数据为e
e = Sample('e', data2['0'].values[0], data2['3'].values[0], data2['4'].values[0], data2['4.1'].values[0], data2['3.1'].values[0])
s5 = Sample('s5', data2['0'].values[1], data2['3'].values[1], data2['4'].values[1], data2['4.1'].values[1], data2['3.1'].values[1])
s6 = Sample('s6', data2['0'].values[2], data2['3'].values[2], data2['4'].values[2], data2['4.1'].values[2], data2['3.1'].values[2])

e, s5, s6
```

```python
def chi_cal(a: Sample, e:Sample, s:Sample, mode):
    global chi_0
    M_lst = 0.24969
    M_hxy = 0.42242
    def chix(dma, dm0, ma, m0):
        return chi_0 * (dma / dm0) * (m0 / ma)
        
    if mode == 3:
        chi = chix(a.m3, s.m3, a.m, s.m)
    elif mode == 4:
        chi = chix(a.m4, s.m4, a.m, s.m)
    elif mode == 31:
        chi = chix(a.m31, s.m31, a.m, s.m)
    elif mode == 41:
        chi = chix(a.m41, s.m41, a.m, s.m)
    else:
        raise ValueError('mode must be 3 or 4 or 31 or 41')
    
    if 'lst' in a.name:
        chi = chi * M_lst
    elif 'hxy' in a.name:
        chi = chi * M_hxy
    return chi
```

```python
# 计算 chi

chi_3 = pd.DataFrame(columns=['name', 'chi_3', 'chi_4', 'chi_31', 'chi_41'])

for i in range(3, 11):
    a = Sample(data2['name'].values[i], data2['0'].values[i], data2['3'].values[i], data2['4'].values[i], data2['4.1'].values[i], data2['3.1'].values[i])
    if '5' in a.name:
        chi_3.loc[i-3] = [a.name, chi_cal(a, e, s5, 3), chi_cal(a, e, s5, 4), chi_cal(a, e, s5, 31), chi_cal(a, e, s5, 41)]
    elif '6' in a.name:
        chi_3.loc[i-3] = [a.name, chi_cal(a, e, s6, 3), chi_cal(a, e, s6, 4), chi_cal(a, e, s6, 31), chi_cal(a, e, s6, 41)]
    else:
        raise ValueError('name must be 5 or 6')

chi_3
```

```python
# 将最终结果逐行转化为latex格式并写入文件
def float_to_latex_sci(num, significant_digits=4):
    # 将数字转化为指定位数的科学计数法字符串
    format_string = "{:." + str(significant_digits-1) + "e}"
    sci_string = format_string.format(num)
    
    # 将科学计数法字符串分割为基数和指数部分
    base, exponent = sci_string.split("e")
    
    # 生成 LaTeX 格式的字符串
    # latex_string = r"${} \times 10^{{{}}}$".format(base, int(exponent))
    latex_string = str(base)
    return latex_string

with open('chi_3.tex', 'w') as f:
    f.write('')
    f.write('\n')

for i in range(chi_3.shape[0]):
    line = chi_3.iloc[i, 1:].values
    latex_line = ''
    for num in line:
        latex = float_to_latex_sci(num)
        latex_line += latex + ' & '

    latex_line = latex_line[:-2] + ' \\\\'
    with open('chi_3.tex', 'a') as f:
        f.write(latex_line)
        f.write('\n')

# 取chi_3的lst5和lst6两行乘以M_lst，取chi_3的hxy5和hxy6两行乘以M_hxy
```

```python
m_error = 8.2e-5
dm_error = 1.2e-4

def r_cal(a:Sample, s:Sample, mode):
    if mode == 3:
        return (a.m3/s.m3) * (a.m/s.m)
    elif mode == 4:
        return (a.m4/s.m4) * (a.m/s.m)
    elif mode == 31:
        return (a.m31/s.m31) * (a.m/s.m)
    elif mode == 41:
        return (a.m41/s.m41) * (a.m/s.m)
    else:
        raise ValueError('ModeError')
    
def r_error_cal(a:Sample, s:Sample, mode):
    def recal(r, ms, ma, dms, dma):
        global m_error, dm_error
        return r*np.sqrt((m_error/ms)**2+(m_error/ma)**2+(dm_error/dms)**2+(dm_error/dma)**2)
    r = r_cal(a, s, mode)
    if mode == 3:
        r_error = recal(r, s.m, a.m, s.m3, a.m3)
    elif mode == 31:
        r_error = recal(r, s.m, a.m, s.m31, a.m31)
    elif mode == 4:
        r_error = recal(r, s.m, a.m, s.m4, a.m4)
    elif mode == 41:
        r_error = recal(r, s.m, a.m, s.m41, a.m41)
    else:
        raise ValueError('ModeError')
    return r_error

r_df = pd.DataFrame(columns=['name', 'r_3', 'r_4', 'r_31', 'r_41'])

for i in range(3, 11):
    a = Sample(data2['name'].values[i], data2['0'].values[i], data2['3'].values[i], data2['4'].values[i], data2['4.1'].values[i], data2['3.1'].values[i])
    if '5' in a.name:
        r_df.loc[i-3] = [a.name, r_cal(a, s5, 3), r_cal(a, s5, 4), r_cal(a, s5, 31), r_cal(a, s5, 41)]
    elif '6' in a.name:
        r_df.loc[i-3] = [a.name, r_cal(a, s6, 3), r_cal(a, s6, 4), r_cal(a, s6, 31), r_cal(a, s6, 41)]
    else:
        raise ValueError('name must be 5 or 6')

r_df
    
```

```python
rer_df = pd.DataFrame(columns=['name', 'er_3', 'er_4', 'er_31', 'er_41'])

for i in range(3, 11):
    a = Sample(data2['name'].values[i], data2['0'].values[i], data2['3'].values[i], data2['4'].values[i], data2['4.1'].values[i], data2['3.1'].values[i])
    if '5' in a.name:
        rer_df.loc[i-3] = [a.name, r_error_cal(a, s5, 3), r_error_cal(a, s5, 4), r_error_cal(a, s5, 31), r_error_cal(a, s5, 41)]
    elif '6' in a.name:
        rer_df.loc[i-3] = [a.name, r_error_cal(a, s6, 3), r_error_cal(a, s6, 4), r_error_cal(a, s6, 31), r_error_cal(a, s6, 41)]
    else:
        raise ValueError('name must be 5 or 6')

rer_df
```

```python
def chi_er_cal(r, r_er):
    global chi_0, er_chi_0
    chi_er = np.sqrt((chi_0 * r_er )**2 + (er_chi_0 * r)**2)
    return chi_er

chi_er_df = pd.DataFrame(columns=['name', 'chi_er_3', 'chi_er_4', 'chi_er_31', 'chi_er_41'])

for i in range(8):
    chi_er_df.loc[i] = [chi_3.iloc[i, 0], 
                        chi_er_cal(r_df.iloc[i, 1], rer_df.iloc[i, 1]), 
                        chi_er_cal(r_df.iloc[i, 2], rer_df.iloc[i, 2]), 
                        chi_er_cal(r_df.iloc[i, 3], rer_df.iloc[i, 3]), 
                        chi_er_cal(r_df.iloc[i, 4], rer_df.iloc[i, 4])]
    if 'hxy' in chi_er_df.iloc[i, 0]:
        chi_er_df.iloc[i, 1:] = chi_er_df.iloc[i, 1:] * M_hxy
    elif 'lst' in chi_er_df.iloc[i, 0]:
        chi_er_df.iloc[i, 1:] = chi_er_df.iloc[i, 1:] * M_lst

chi_er_df
```

```python
def to_significant_digits(value, digits):
    """
    Convert a float in scientific notation (or normal notation) to its normal representation 
    with a given number of significant digits.

    Parameters:
    value (float): Input value in scientific notation.
    digits (int): Number of significant digits required.

    Returns:
    str: Value represented with the required number of significant digits.
    """
    
    # Convert the value to string in scientific notation
    s = "{:.{digits}e}".format(value, digits=digits-1)
    mantissa, exponent = s.split('e')
    return mantissa
```

```python
# 将chi_er_df转化为latex
with open('chi_er.tex', 'w') as f:
    f.write('')
    f.write('\n')

for i in range(chi_er_df.shape[0]):
    line = chi_er_df.iloc[i, 1:].values
    latex_line = ''
    for num in line:
        try:
            latex = to_significant_digits(num, 4)
            latex_line += latex + ' & '
        except:
            latex_line += str(num) + ' & '

    latex_line = latex_line[:-2] + ' \\\\'
    with open('chi_er.tex', 'a') as f:
        f.write(latex_line)
        f.write('\n')
```
