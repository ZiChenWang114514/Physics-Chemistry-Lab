```python
import numpy as np
import matplotlib.pyplot as plt

# 定义透光率 T 的范围（避免 T=0 的情况，因为会导致分母为零）
T = np.linspace(0.01, 1.0, 1000)  # 从 0.01 到 1.0 的 1000 个点

# 计算相对误差 E_r
E_r = 0.434 * (0.01/ (T * np.log(T)))

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(T, E_r, label=r'$E_r = 0.434 \frac{d T}{T \ln T}$')
plt.xlabel('透光率 T')
plt.ylabel('相对误差 E_r')
plt.title('透光率对相对误差的影响')
plt.legend()
plt.grid(True)
plt.show()
```
