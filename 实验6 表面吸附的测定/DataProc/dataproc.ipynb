```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 输入数据


# 读取数据
df1 = pd.read_csv('data.csv', sep=' ', header=None)
df1
```

```python
# 创建df column name = ['$c_{n-ce{BuOH}}/\si{mol\cdot L^{-1}}$', '$h_{0,l}$/\si{cm}', '$h_{0,r}$/\si{cm}', '$h_{1,l}$/\si{cm}', '$h_{1,r}$/\si{cm}', '$\Delta h$/\si{cm}']
df1.columns = ['$c_{nce{BuOH}}/\si{mol\cdot L^{-1}}$', '$h_{0,l}$/\si{cm}', '$h_{0,r}$/\si{cm}', '$h_{1,l,1}$/\si{cm}', '$h_{1,l,2}$/\si{cm}', '$h_{1,l,3}$/\si{cm}', '$h_{1,r,1}$/\si{cm}', '$h_{1,r,2}$/\si{cm}', '$h_{1,r,3}$/\si{cm}']
df1.head()
```

```python
# 在df1的最后添加一行，值为np.log10(df1.iloc[:, 0])
df1['$\ln c_{n\ce{BuOH}}$'] = np.log(df1.iloc[1:, 0])
df1['$\bar{h}_{1,l}$/\si{cm}'] = df1.iloc[:, 3:6].mean(axis=1)
df1['$\bar{h}_{1,r}$/\si{cm}'] = df1.iloc[:, 6:9].mean(axis=1)
df1['$\Delta h$/\si{cm}'] = (df1['$\bar{h}_{1,r}$/\si{cm}'] - df1["$h_{0,r}$/\si{cm}"]) - (df1['$\bar{h}_{1,l}$/\si{cm}'] - df1['$h_{0,l}$/\si{cm}'])
df1.head()
```

```python


# 假设 df1 已经存在，并且有一个列 '$\Delta h$/\si{cm}'

gamma = [71.18]  # 初始 gamma 值列表
for i in range(1, df1.shape[0]):
    gamma.append(df1['$\Delta h$/\si{cm}'][i] * gamma[0] / df1['$\Delta h$/\si{cm}'][0])

gamma_error = [0]  # 初始 gamma_error 值列表
def g_err(dh, dh0):
    return np.sqrt(1 + np.square(dh / dh0)) * 71.18 * 0.047 / dh0

for i in range(1, df1.shape[0]):
    gamma_error.append(g_err(df1['$\Delta h$/\si{cm}'][i], df1['$\Delta h$/\si{cm}'][0]))

# 将 gamma 和 gamma_error 结合起来，并添加到 df1
df1['$\gamma$/\si{mN\cdot m^{-1}}'] = [f"{gamma[0]:.2f}"] + [f"{g:.2f} $\pm$ {ge:.2f}" for g, ge in zip(gamma[1:], gamma_error[1:])]

df1
```

```python
# 将df1改变列顺序为:$c_{n-ce{BuOH}}/\si{mol\cdot L^{-1}}$ $h_{0,l}$/\si{cm} $h_{1,l,1}$/\si{cm} $h_{1,l,2}$/\si{cm} $h_{1,l,3}$/\si{cm} $\bar{h}_{1,l}$/\si{cm} $h_{0,r}$/\si{cm} $h_{1,r,1}$/\si{cm} $h_{1,r,2}$/\si{cm} $h_{1,r,3}$/\si{cm} $\bar{h}_{1,r}$/\si{cm} $\Delta h$/\si{cm}
df1_1 = df1.iloc[:, [0, 1, 3, 4, 5, 10, 2, 6, 7, 8, 11, 12]].copy()
df1_1.head()
```

```python
# 将第一列之后的数据保留两位小数
# 将第一列改为字符
df1_1.iloc[:, :1] = df1_1.iloc[:, :1].map(lambda x: '%.3f' % x)
df1_1.to_latex('data1.tex', index=False, float_format='%.2f')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.ticker import FuncFormatter
from scipy.optimize import curve_fit
from scipy.stats import linregress
from scipy.interpolate import splrep, splev

# 自定义格式器
def format_x(value, tick_number):
    return f'{value:.3f}'

def format_y(value, tick_number):
    return f'{value:.2f}'
        
# 设置 matplotlib 参数
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Times New Roman'
plt.rcParams['mathtext.it'] = 'Times New Roman:italic'
plt.rcParams['mathtext.bf'] = 'Times New Roman:bold'
plt.rcParams['font.size'] = 16
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16


# 假设 df4 已经存在并包含所需的列
df4 = df4.astype(str)
df4['$\\Delta\\gamma$'] = df4['$\\gamma$/\\si{mN\\cdot m^{-1}}'].map(lambda x: x[-4:])
df4['$\\gamma$/\\si{mN\\cdot m^{-1}}'] = df4['$\\gamma$/\\si{mN\\cdot m^{-1}}'].map(lambda x: x[:4])

# 转换为 float 类型
df4 = df4.astype(float)

# 线性拟合函数
def linear_model(x, a, b):
    return a * x + b

# 使用B-spline对数据进行平滑
x_data = df4.iloc[:, 0]
y_data = df4.iloc[:, 1]
y_error = df4.iloc[:, 2]
spl = splrep(x_data, y_data)

x_spline = np.linspace(x_data.min(), x_data.max(), 300)
y_spline = splev(x_spline, spl)

# 特定的c值
c_values = [0.15, 0.20, 0.30, 0.40, 0.45]
log_c_values = np.log(c_values)

# 在B-spline曲线上找到特定点及其误差
selected_y = splev(log_c_values, spl)
selected_y_error = y_error[np.searchsorted(x_data, log_c_values)]  # 近似误差

# 保存到新的DataFrame
df_selected = pd.DataFrame({
    'log_c': log_c_values,
    'gamma': selected_y,
    'gamma_error': selected_y_error
})

# 对选定点进行线性拟合
popt, pcov = curve_fit(linear_model, df_selected['log_c'], df_selected['gamma'], sigma=df_selected['gamma_error'])
a_fit, b_fit = popt
a_error, b_error = np.sqrt(np.diag(pcov))

# 计算 R^2 值
slope, intercept, r_value, _, _ = linregress(df_selected['log_c'], df_selected['gamma'])

# x 轴范围为整个数据集的范围
x_range = np.linspace(x_data.min(), x_data.max(), 300)

# 可视化
plt.figure(figsize=(8, 6), dpi=1200)

# 主轴（实际只有一套轴，但网格线将具有主网格线和副网格线）
ax = plt.gca()
ax.errorbar(x_data, y_data, yerr=y_error, fmt='o', color='blue', label=r'$\gamma$ with error')
ax.plot(x_spline, y_spline, color='grey', alpha=0.5, label='B-spline')  # 绘制B-spline曲线
ax.errorbar(df_selected['log_c'], df_selected['gamma'], yerr=df_selected['gamma_error'], fmt='s', color='orange', label='Selected points for fitting')
ax.plot(x_range, linear_model(x_range, a_fit, b_fit), color='red', label='Linear fit')
ax.xaxis.set_major_formatter(FuncFormatter(format_x))
ax.yaxis.set_major_formatter(FuncFormatter(format_y))
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())


# 设置主网格线
ax.grid(which='major', alpha=0.5)

# 设置副网格线
ax.grid(which='minor', alpha=0.2)

# 设置标签和图例
ax.set_xlabel(r'$\ln (c_{n\mathrm{BuOH}}/\mathrm{mol\cdot L^{-1}})$')
ax.set_ylabel(r'$\gamma\:/\mathrm{mN\cdot m^{-1}}$')

ax.legend()

plt.show()

# 输出回归直线的表达式、参数误差和 R^2
print(f"回归直线: y = {a_fit:.2f}x + {b_fit:.2f}")
print(f"参数误差: a_error = {a_error:.2f}, b_error = {b_error:.2f}")
print(f"R^2 = {r_value**2:.6f}")
df_selected['c'] = np.exp(df_selected['log_c'])
df_selected = df_selected[['c', 'log_c', 'gamma', 'gamma_error']]
print(df_selected.to_latex(index=False, float_format='%.3f'))
```
